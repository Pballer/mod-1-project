```python
tn_movie_budgets = pd.read_csv('./data/tn.movie_budgets.csv.gz')
```

```python
tn_movie_budgets.head()
```

```python
tn_movie_budgets.release_date = pd.to_datetime(tn_movie_budgets.release_date)
```

```python
tn_movie_budgets['release_year'] = tn_movie_budgets.release_date.dt.year
```

```python
tn_movie_budgets.loc[:, 'production_budget':'worldwide_gross'] = \
    tn_movie_budgets.loc[:, 'production_budget':'worldwide_gross'].replace('[^.0-9]', '', regex=True).astype(int)
```

```python
tn_movie_budgets['dom_roi'] = (((tn_movie_budgets.domestic_gross - tn_movie_budgets.production_budget) / tn_movie_budgets.production_budget) * 100).astype(int)
```

```python
tn_movie_budgets['world_roi'] = (((tn_movie_budgets.worldwide_gross - tn_movie_budgets.production_budget) / tn_movie_budgets.production_budget) * 100).astype(int)
```

```python
tn_movie_budgets.head()
```

```python
#James Cameron not found b/c imdb is 2010+. Avatar is 2009.  Would need to pull older data.
imdb_basics_rating[imdb_basics_rating.primary_title == 'Avatar']
```

```python
#join imdb with tn
imdb_tn_primary_title = pd.merge(
    imdb_basics_rating, 
    tn_movie_budgets,
    left_on=['primary_title', 'start_year'],
    right_on=['movie', 'release_year'],
    how='inner',
)
```

```python
imdb_tn_original_title = pd.merge(
    imdb_basics_rating, 
    tn_movie_budgets,
    left_on=['original_title', 'start_year'],
    right_on=['movie', 'release_year'],
    how='inner',
)
```

```python
imdb_tn_all = pd.concat([imdb_tn_primary_title, imdb_tn_original_title])
```

```python
imdb_tn_all.drop_duplicates(inplace=True)
```

```python
imdb_tn_all['main_genre'] = imdb_tn_all.genres.str.split(',', expand=True)[0].astype('category')
```

```python
imdb_tn_all.main_genre.unique()
```

```python
imdb_tn_all.groupby('main_genre')[['domestic_gross', 'production_budget']].corr().iloc[0::2,-1]
```

```python
genre_dom_gross_corr_prod_budget = imdb_tn_all.groupby('main_genre')[['domestic_gross', 'production_budget']].corr()
```

```python
genre_dom_gross_corr_prod_budgest = genre_dom_gross_corr_prod_budget.iloc[::2, -1].reset_index()
```

```python
genre_dom_gross_corr_prod_budget.drop(columns='level_1', inplace=True)
```

```python
genre_dom_gross_corr_prod_budget.rename(columns={'production_budget': 'prod_dom_gross_corr'}, inplace=True)
```

```python
#Low budget with high domestic gross
genre_dom_gross_corr_prod_budget.sort_values(by='prod_dom_gross_corr')
```
